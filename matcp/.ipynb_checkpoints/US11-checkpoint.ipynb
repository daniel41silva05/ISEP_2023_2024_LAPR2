```python
import pandas as pd
import matplotlib.pyplot as plt

# Leitura do arquivo CSV
df = pd.read_csv("C:/Users/danie/Desktop/UNI/2º Semestre/Projeto/datasets matcp/simulacao_ficheiro_us11.csv", delimiter=';')

# Indicação do tipo de cada uma das três variáveis
print("Tipos de variáveis:")
print("Idade:")
print(df['Age'].value_counts())
print("\nRecomendaria o parque para outros:")
print(df['Recommend'].value_counts())
print("\nFrequência de visitas ao parque por mês:")
print(df['Visit park'].describe())

# Proporção de usuários de cada faixa etária que recomendaria o parque
recommendation_counts = df.groupby('Age')['Recommend'].value_counts(normalize=True)[:, 'Y']
print("\nProporção de usuários de cada faixa etária que recomendaria o parque:")
print(recommendation_counts)

# Criação de boxplot para cada faixa etária, referente à frequência mensal de uso do parque
plt.figure(figsize=(8, 6))
df.boxplot(column='Visit park', by='Age', grid=False)
plt.title('Boxplot da Frequência Mensal de Visita ao Parque por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Frequência Mensal de Visita ao Parque')
plt.suptitle('')
plt.show()
```
